```
#dataset = Lreviewdata.select("Text").limit(50)
dataset = Lreviewdata.filter(Lreviewdata.productname == getArgument("SelectedProductname")).select("Text").limit(50)
text=dataset.select('Text').rdd.map(lambda x: x[0]).collect()
text = ' '.join(text)
```

```
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import matplotlib.pyplot as plt
def show_wordcloud(text):
    wordcloud = WordCloud(
        background_color='white',
        stopwords=set(STOPWORDS),
        max_words=200,
        max_font_size=40, 
        scale=3,
        random_state=1
    ).generate(str(text))
    fig = plt.figure(1, figsize=(40, 20))
    plt.axis('off')
    plt.imshow(wordcloud)
    plt.show()
    display()
```

```
show_wordcloud(text)
```
