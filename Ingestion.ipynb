```
#convert data files to strict json
import json
import gzip

def parse(path):
  g = gzip.open(path,'r')
  for l in g:
    yield json.dumps(eval(l))
f = open("/tmp/cellphone_reviews.json","w")
for l in parse("/dbfs/mnt/rawsource/amazon/unlabeled/reviews_Cell_Phones_and_Accessories_5.json.gz"):
  f.write(l + '\n')

dbutils.fs.mv("file:///tmp/cellphone_reviews.json","/mnt/rawsource/amazon/unlabeled/cellphone_reviews.json")
```
